This notebook will merge all the translation scores for the various models evaluated and provide mean scores overall

```python
import os
import glob
import pandas as pd

# The directory to where csv score files generated by the translator_pipeline.py were written to
output_dir = ''
os.chdir(output_dir)
target_language = '' #  e.g. Set to fr for French/ es for Spanish
```

```python
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
# Combine all files in the list
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])
# Export to csv
combined_csv.to_csv(target_language + "_combined_csv.csv", index=False, encoding='utf-8-sig')
```
